```python
import pandas as pd

data = {'luas' : [6, 8, 10, 14, 18],
        'harga' : [7, 9, 13, 17.5, 18]}

df = pd.DataFrame(data)
print (df)
#membuat tabel dengan nama kolomm luas dan harga
```

```python
import matplotlib.pyplot as plt

df.plot(kind='scatter', x='luas', y='harga')
plt.title('Perbandingan Luas Tanah dengan Harga Tanah')
plt.xlabel('Luas Tanah (m persegi)')
plt.ylabel('Harga Tanah (dollar)')
plt.xlim(0,25)
plt.ylim(0,25)
plt.grid(True)
plt.show()
# untuk menampilkan perbandingan antara luas tanah dengan harga tanah menggunakan scatter plot
```

```python
import numpy as np
x=np.array(df['luas'])
y=np.array(df['harga'])

print(f'X : {x}')
print(f'y : {y}')
# mengubah data yang sebelumnya ditabel menjadi array numpy
```

```python
X = x.reshape (-1,1)

print(f'X shape: {X.shape}')
print(f'X: {X}')
# mencetak berapa baris dan kolom datanya serta mengubah data yang awalnya array x menjadi array X atau array 2D
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X,y)
#unutk melatih model regresi linear
```

```python
X_vis = np.array([0,25]).reshape(-1,1)
y_vis = model.predict(X_vis)
# untuk membuat array pada sumbu x dan memprediksi harga menggunakan model regresi pada sumbu y
```

```python
plt.scatter(X,y)
plt.plot(X_vis,y_vis, color='red')

plt.title('Perbandingan Luas Bangun dan Harga')
plt.xlabel('Luas Tanah (m persegi)')
plt.ylabel('Harga Tanah (dollar)')
# plt.xlim(0,25)
# plt.ylim(0,25)
plt.grid(True)
plt.show()
#unutm menampilkan perbandingan antara luas bangun dan harga tanah menggunakan scatter plot
```

```python
print(f'intercept : {model.intercept_}')
print(f'slope : {model.coef_}')
# menampilkan parameter regresi linear menggunakan intercept dan slope
```

```python
luas_tanah = np.array([12,15,20]).reshape(-1,1)
print(luas_tanah.shape)
luas_tanah
#berisikan array dengan 3 baris 1 kolom yang menampilkan luas tanah
```

```python
harga_prediksi = model.predict(luas_tanah)
harga_prediksi
#memprediksi harga tanah dengan model regresi linear
```

```python
for luas, hrg in zip(luas_tanah, harga_prediksi): print (f'Luas Tanah {luas}: harga prediksi: {hrg}')
#menampilkan hasil prediksi harga tanah
```

```python
X_train = np.array([6, 8, 10, 14, 18]).reshape(-1,1)
y_train = np.array([7, 9, 13, 17.5, 18])

X_test = np.array([9, 9, 11, 16, 12]).reshape(-1,1)
y_test = np.array([11, 8.5, 15, 18, 11])
#menyiapkan data peatihan dan data uji
```

```python
model=LinearRegression()
model.fit(X_train,y_train)
#untuk melatih model regresi linear
```

```python
from sklearn.metrics import r2_score

y_pred = model.predict(X_test)
r_squared = r2_score(y_test, y_pred)
print(f'R-squared: {r_squared}')
#mengukur seberapa baik model menjelaskan data
```
